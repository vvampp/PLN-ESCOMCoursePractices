```python
import re
import pandas as pd
from collections import Counter
```

```python
tweets = pd.read_csv("tweets.csv")
```

```python
column = tweets['text']
```

```python
print(tweets)
```

# Busquedas Regex

## Hashtags

```python
hashtagRe = r'#\w+'
```

```python
hashtagCounter = 0
hashtagAcc = {}
```

```python
for idx, row in enumerate(column): 
    matches = re.findall(hashtagRe,row)
    if matches:
        print(idx +1,matches)
    for match in matches:
        hashtagCounter += 1
        if match in hashtagAcc:
            hashtagAcc[match] += 1
        else:
            hashtagAcc[match] = 1
```

```python
sortedHD =  dict(sorted(hashtagAcc.items(), key=lambda item: item[1], reverse=True))
print(sortedHD)
```

```python
print(hashtagCounter)
```

## Users

```python
userRe = r'@[A-Za-z_]\w{0,14}'
```

```python
userCounter = 0
userAcc = {}
```

```python
for idx, row in enumerate(column):
    matches = re.findall(userRe,row)
    if matches:
        print(idx+1,matches)
    for match in matches:
        userCounter += 1
        if match in userAcc:
            userAcc[match] += 1
        else:
            userAcc[match] = 1
```

```python
sortedUD =  dict(sorted(userAcc.items(), key=lambda item: item[1], reverse=True))
print(sortedUD)
```

```python
print(userCounter)
```

## URLs

```python
urlRe = re.compile(
    r'https?:\/\/(?:(?:(?:[A-Za-z0-9-])+\.))+(?:[A-Za-z]){2,}(?::\d{1,5})?(?:(?:\/(?:[\w\-\.~%])+)*)?(?:\?(?:(?:[\w\-\.~%])+=(?:[\w\-\.~%])*(?:&(?:[\w\-\.~%])+=(?:[\w\-\.~%])*)*)?)?(?:#(?:[\w\-\.~%\/\+\!\@\(\)\[\]\{\}]+))??')
```

```python
urlCounter = 0
urlAcc = {}
```

```python
for idx, row in enumerate(column): 
    matches = re.findall(urlRe,row)
    if matches:
        print(idx+1,matches)
    for match in matches:
        urlCounter += 1
        if match in urlAcc:
            urlAcc[match] += 1
        else:
            urlAcc[match] = 1
```

```python
sortedURLD =  dict(sorted(urlAcc.items(), key=lambda item: item[1], reverse=True))
print(sortedURLD)
```

```python
print(urlCounter)
```

## Horas

```python
horaRe = r'(?:(?:(?:\d(?<=[0-2])\d)|(?:[1-9]))(?::(?:[0-5]\d))+?(?:am|AM|pm|PM|hrs?|HRS?|horas?)?)|(?:\d(?<=[0-2])\d|[1-9])(?::[0-5]\d)?\s?(?:am|AM|pm|PM|hrs?|HRS?horas?)'
```

```python
horaCounter = 0
horaAcc = {}
```

```python
for idx, row in enumerate(column): 
    matches = re.findall(horaRe,row)
    if matches:
        print(idx+1,matches)
    for match in matches:   
        horaCounter += 1
        if match in horaAcc:
            horaAcc[match] += 1
        else:
            horaAcc[match] = 1
```

```python
sortedHoraD =  dict(sorted(horaAcc.items(), key=lambda item: item[1], reverse=True))
print(sortedHoraD)
```

```python
print(horaCounter)
```

## Emoticonos

```python
emoticonoRe = r'(?:>?(?::|;)-?(?:\)|\(|D|p|P|0|\\|O|\|))|(?:(?:x|X)D)'
```

```python
emoticonoCounter = 0
emoticonoAcc = {}
```

```python
for idx, row in enumerate(column): 
    matches = re.findall(emoticonoRe,row)
    if matches:
        print(idx+1,matches)
    for match in matches:   
        emoticonoCounter += 1
        if match in emoticonoAcc:
            emoticonoAcc[match] += 1
        else:
            emoticonoAcc[match] = 1
```

```python
sortedEmoticonoD =  dict(sorted(emoticonoAcc.items(), key=lambda item: item[1], reverse=True))
print(sortedEmoticonoD)
```

```python
print(emoticonoCounter)
```

## Emojis

```python
emojiRe = re.compile(
    r'(?:'
    r'[\U0001F600-\U0001F64F'  # Emoticons
    r'\U0001F300-\U0001F5FF'   # S√≠mbolos y pictogramas varios
    r'\U0001F680-\U0001F6FF'   # Transporte y mapas
    r'\U0001F1E0-\U0001F1FF'   # Banderas
    r'\U00002702-\U000027B0'   # S√≠mbolos varios
    r'\U000024C2-\U0001F251'   # Otros s√≠mbolos adicionales
    r'\U0001F900-\U0001F9FF'   # Emojis de 2017 en adelante
    r'\U0001FA70-\U0001FAFF'   # S√≠mbolos adicionales
    r'\U00002600-\U000026FF'   # S√≠mbolos miscel√°neos
    r'\U00002B50-\U00002B55'   # Estrellas
    r']|'
    r'\U0001F1E6-\U0001F1FF'   # Banderas de pa√≠ses
    r')(?![\uFE0F\u200D\u1F3FB-\u1F3FF])'  # Ignorar FE0F, ZWJ, y modificadores de tono de piel
)
```

```python
emojisCounter = 0
emojisAcc = {}
```

```python
for idx, row in enumerate(column): 
    matches = re.findall(emojiRe,row)
    if matches:
        print(idx+1,matches)
    for match in matches:   
        emojisCounter += 1
        if match in emojisAcc:
            emojisAcc[match] += 1
        else:
            emojisAcc[match] = 1
```

```python
sortedEmojiD =  dict(sorted(emojisAcc.items(), key=lambda item: item[1], reverse=True))
print(sortedEmojiD)
```

```python
print(emojisCounter)
```

# Reporte CSV

```python
diccionarios = [sortedHD,sortedUD,sortedURLD,sortedHoraD,sortedEmoticonoD,sortedEmojiD]
```

```python
tabla = {
    'String': ['Hashtags','Usuarios', 'URLs', 'Horas', 'Emoticonos', 'Emojis'],
    'Total Frequency': [hashtagCounter,userCounter,urlCounter,horaCounter,emoticonoCounter,emojisCounter],
    'Top 10': diccionarios
}
```

```python
df = pd.DataFrame(tabla)
df['Top 10'] = df['Top 10'].apply(lambda x: '\n'.join([f'{repr(k)}: {v}' for k, v in list(x.items())[:10]]))
```

```python
df.to_csv('tablaPract1.csv', index=False, encoding='utf-8')
```
